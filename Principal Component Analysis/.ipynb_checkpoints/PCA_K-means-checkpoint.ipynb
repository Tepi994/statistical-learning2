```python
%matplotlib inline
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.datasets import fetch_openml

from sklearn.decomposition import PCA
```

```python
def mostrar_digito(digito):
    digito = digito.reshape(8, 8)

    plt.imshow(digito, cmap = matplotlib.cm.binary,
               interpolation="nearest")
    plt.axis("off")
    plt.show()
```

# Loading Data (MNIST)

```python
from sklearn.datasets import load_digits
digits = load_digits()

x = digits.data
x.shape
```

```python
digits.target
```

```python
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(x)
```

```python
#### PCA 2 dimensions ######
x_norm = x/255
pca = PCA(n_components= 2)
principalComponents = pca.fit_transform(x)
X2D = pd.DataFrame(data = principalComponents, 
                    columns = ['pca_1','pca_2'])
X2D = principalComponents
```

```python
print(x.shape)
print(X2D.shape)
```

```python
plt.scatter(X2D[:, 0], X2D[:, 1])
plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python
plt.scatter(X2D[:, 0], X2D[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.colorbar()
```

## K-means Clustering using Sklearn

[Documentacion oficial](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```python
## K-means
kmeans = KMeans(n_clusters=10)
Km = kmeans.fit(X2D)
```

```python
clusters = kmeans.predict(X2D)
```

```python
kmeans.cluster_centers_.shape
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X2D[:, 0], X2D[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
